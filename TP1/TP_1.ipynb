Question1

```python

df_cities.columns = df_cities.columns.str.replace('"', '').str.strip()


df_cities["Latitude_total"]  = df_cities["LatD"] + df_cities["LatM"]/60 + df_cities["LatS"]/3600
df_cities["Longitude totale "]  = df_cities["LonD"] + df_cities["LonM"]/60 + df_cities["LonS"]/3600


```

Question 2:Filterer les villes situé au nord

```python
df_cities = df_cities[df_cities["NS"] == ' "N"']
df_cities
```

Question3: fusionner les data frame

```python
df_merged= pd.merge(df_cities,df_populations, how='left', on='City')
df_merged["City"]

```

```python
print(df_merged.groupby('City').size())
```

 
Question 4 Calculer la population projetée pour 2025

```python
df_merged_grouped= df_merged
df_merged_grouped["Population_2025"] = df_merged_grouped["Population_2020"] * (1 + df_merged_grouped["Growth_rate"] / 100) ** 5
df_merged_grouped
```

Question5: afficher les villes avec une population projetée supérier a 1 million

```python
df_merged_grouped_1millon=df_merged_grouped[df_merged_grouped["Population_2025"]>1_000_000]
df_merged_grouped_1millon[["City", "Population_2025"]]
```

Question 10 Trouver la moyenne et la médiane de la population des villes

```python
moyenne_population = df_merged_grouped["Population_2025"].mean()


mediane_population = df_merged_grouped["Population_2025"].median()
```

Question 11   Créer une nouvelle colonne pour la population normalisée 

```python

pop_min = df_merged_grouped["Population_2025"].min()
pop_max = df_merged_grouped["Population_2025"].max()

df_merged_grouped["Population_Normalisé"] = (df_merged_grouped["Population_2025"] - pop_min) / (pop_max - pop_min)


df_merged_grouped.head()
```
